```python
import sympy 
import numpy as np
from sympy.plotting import plot
```

```python
x = sympy.symbols('x')
sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0)
```

 1. Определить корни

```python
sympy.solveset(sympy.Eq(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, 0), x)
```

Функция имеет бесконечное число интервалов, где она возрастает или убывает. Вершину определить можно только на каком-то заданном интервале 

График функции

```python
plot(-12*x**4*sympy.sin(sympy.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30, (x, -100, 100, 0.0001))
```

```python
f = -12*x**4 * (np.sin(np.cos(x))) - 18*x**3 + 5*x**2 + 10*x - 30

positiv_lst = []
negativ_lst = []

for i in np.arange(-100, 101):
    result = -12*i**4 * (np.sin(np.cos(i))) - 18*i**3 + 5*i**2 + 10*i - 30
    if result > 0:
        positiv_lst.append(i)
    else:
        negativ_lst.append(i)

print(f'Промежутки, на которых f > 0, в интервале от -100 до 100: {positiv_lst[-1]} --> {positiv_lst[0]}')
print(positiv_lst)
print(f'Промежутки, на которых f < 0, в интервале от -100 до 100:{negativ_lst[0]} --> {negativ_lst[-1]}')
print(negativ_lst)
```
